```python
from datetime import date
from datetime import timedelta
import requests
import json


def get_vacancy_idxs(idxs, prole, dateFrom, dateTo):
    url = "https://api.hh.ru/vacancies"

    headers = {
        'User-Agent': "PostmanRuntime/7.15.0",
        'Accept': "*/*",
        'Cache-Control': "no-cache",
        'Postman-Token': "e9113f34-84fc-4028-aa10-08001de3769a,7bc059ec-8fcc-4ef8-8728-70f69f7c68af",
        'Host': "api.hh.ru",
        'accept-encoding': "gzip, deflate",
        'Connection': "keep-alive",
        'cache-control': "no-cache"
    }
    for i in range(0, 20):
        querystring = {"industry": "7", "professional_role": prole, "date_from": dateFrom, "date_to": dateTo,
                       "per_page": "100", "page": i}
        response = requests.request("GET", url, headers=headers, params=querystring)
        data = response.json()
        if 'items' in data:
            for el in data['items']:
                idxs.add(el['id'])
        else:
            return
        pages = int(data["pages"])
        if i >= pages:
            return

        
```

```python
currentDate = date.today()
proles = {156, 160, 10, 12, 150, 25, 165, 34, 36, 73, 155, 96, 164, 104, 157, 107, 112, 113, 148, 114, 116, 121, 124,
          125, 126}
idxs = set()
for p in proles:
    for d in range(0, 29):
        dateTo = currentDate - timedelta(days=d)
        dateFrom = dateTo - timedelta(days=1)
        get_vacancy_idxs(idxs, p, dateFrom, dateTo)
print(len(idxs))
```

```python
import re
def remove_tags(text):
    return re.sub(r'\<.*\>', '', text)
```

```python
import pandas as pd
import time
data = []
data = pd.DataFrame(
    columns=['domain_id', 'domain_name', 'profearea_id', 'profearea_name', 'ad_id', 'name', 'description', 'req',
             'resp', 'cond', 'date'])
k=0;
for idx in idxs:
    k=k+1
    if k%200==0:
        time.sleep(0.2)
    url = "https://api.hh.ru/vacancies/" + str(idx)
    print(url)
    headers = {
        'User-Agent': "PostmanRuntime/7.15.0",
        'Accept': "*/*",
        'Cache-Control': "no-cache",
        'Postman-Token': "c16c7cdf-120d-49f7-ba2d-a06d78a3f8f2,384c178b-a538-4c54-ae90-ab4817e9486e",
        'Host': "api.hh.ru",
        'accept-encoding': "gzip, deflate",
        'Connection': "keep-alive",
        'cache-control': "no-cache"
    }
    response = requests.request("GET", url, headers=headers)
    response = json.loads(response.text)
    desc = response['description']
    date = response['published_at']

    try:
        spec = response['professional_roles']
        profarea_id = []
        profarea_name = []
        for el in spec:
            profarea_id.append(el['id'])
            profarea_name.append(el['name'])

        desc_arr = desc.split('<ul>')
        if (len(desc_arr) == 4 or len(desc_arr) == 3):
            resp = desc_arr[1].split('</ul>')[0].replace('</li>', '').split('<li>')[1:]
            for el in resp:
                el = remove_tags(el)
            req = desc_arr[2].split('</ul>')[0].replace('</li>', '').split('<li>')[1:]
            for el in req:
                el = remove_tags(el)
            if len(desc_arr) == 4:
                cond = desc_arr[3].split('</ul>')[0].replace('</li>', '').split('<li>')[1:]
                for el in cond:
                    el = remove_tags(el)
            else:
                cond = None
            for i in range(len(profarea_id)):
                new_row_data = {
                    'domain_id': '7', 'domain_name': ' , , ',
                    'profearea_id': profarea_id[i], 'profearea_name': profarea_name[i], 'ad_id': idx,
                    'name': response['name'], 'description': desc, 'req': req, 'resp': resp,
                    'cond': cond, 'date': date
                }
                new_row = pd.DataFrame([new_row_data])
                data = pd.concat([data, new_row], ignore_index=True)
    except:
        print('oops')
```

```python
def elim_com(x):
    try:
        x=x.replace('<p>','')
        x=x.replace('</p>','')
        x=x.replace('<em>','')
        x=x.replace('</em>','')
        return x
    except ValueError:
        return x
```

```python
data_fp['conditions']=data_fp['conditions'].apply(elim_com)
data_fp['requirements']=data_fp['requirements'].apply(elim_com)
data_fp['responsibilities']=data_fp['responsibilities'].apply(elim_com)
```
